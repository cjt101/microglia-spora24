```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp_data = pd.read_excel('Temperature, Time to Clear (magnitude=10).xlsx', header=17)
temp_data.head()
```

```python
temp_data = temp_data.filter(['temperature', 'ticks'])
temp_data.rename(columns={'ticks':'Ticks'}, inplace=True)
temp_data.head()
```

```python
# don't know if this data is complete, so checking to see all groups
# have the same number of trials
print(temp_data.groupby(['temperature']).size().to_string())
```

```python
temp_data.groupby('temperature').mean()
```

```python
plt.style.use('bmh')
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.dpi'] = 300

(
    temp_data
    .groupby('temperature')
    .mean()
    .plot()
)

plt.legend(['Average over 10 trials'])
#plt.title('Average Time Taken to Clear Damaged Neurons (magnitude * 10)', pad=20)
plt.xlabel('temperature (C\xb0)')
plt.ylabel('ticks')
plt.ylim(700, 1400)
plt.xlim(35.9, 40.1)

plt.savefig('temp-ver-1-1.png', bbox_inches='tight')
```

```python
# alternate graph where y-axis starts at 0 instead of 700
(
    temp_data
    .groupby('temperature')
    .mean()
    .plot()
)

plt.legend(['Average over 10 trials'])
#plt.title('Average Time Taken to Clear Damaged Neurons (magnitude * 10)', pad=20)
plt.ylabel('ticks')
plt.xlabel('temperature (C\xb0)')
plt.ylim(0, 1400)
plt.xlim(35.9, 40.1)

# Uncomment to save
# plt.savefig('temp-ver-2-1.png', bbox_inches='tight')
```

```python
eat_data = pd.read_excel('Eat Probability, Time to Clear.xlsx', header=23)
eat_data.head()
```

```python
# weird data duplication happened, removing rows that have
# no value for eat-probability
eat_data = eat_data[eat_data['eat-probability'].notnull()]
eat_data.reset_index(inplace=True)
eat_data.rename(columns={'Unnamed: 1':'Ticks'}, inplace=True)
eat_data = eat_data.filter(['eat-probability', 'Ticks'])

eat_data.head()
```

```python
# eat_data.head(51)
```

```python
# don't know if this data is complete, so checking to see all groups
# have the same number of trials
# print(eat_data.groupby(['eat-probability']).size().to_string())
```

```python
# get min/max for ylabel ranges
print(eat_data.groupby('eat-probability').mean().max())
print(eat_data.groupby('eat-probability').mean().min())
```

```python
(
    eat_data
    .groupby('eat-probability')
    .mean()
    .plot()
)

plt.legend(['Average over 50 trials'])
plt.ylabel('ticks')
plt.xlabel('eat probability (%)')
plt.ylim(650, 8000)
plt.xlim(10.0, 100.1)
plt.xticks(np.arange(0, 101, 10))

plt.savefig('eat-prob.png', bbox_inches='tight')
```
